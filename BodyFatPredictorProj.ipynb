## Feature Engineering

##### Fat Distribution and Body Type Feature (ratios):

One of the things we wanted to focus on in our project is not only being able to predict body fat percentages for individuals based of their fat distribution but understanding what insights we can gain from the type of fat distribution. Established research emphasizes that body fat distribution is a better predictor of the health hazards of obesity than is the absolute amount of body fat. Individuals with an upper body fat pattern, reflecting an excess of intra-abdominal or visceral fat, have significantly greater risk for diabetes, hypertension, hypertriglyceridemia, ischemic heart disease, some cancers, and death from all causes [J. Stevens, K.P. Truesdale, in Encyclopedia of Human Nutrition (Second Edition), 2005].

This is why we decided to perform some feature engineering and add to our dataset, focusing on adding a feature consisting of ratios of fat depositon between body parts such as chest-to-hip and abdomen-to-hip (also knowns as waist-to-hip or WHR). These will help help identifying the kinds of body types and help understand what health conditions entail those body types.

```python
# Creating new features representing ratios between different body parts
# We selected these ratios as according to the research we referenced, these are good indicators for metrics such as obesity
data['Chest_to_Hip_Ratio'] = data['Chest'] / data['Hip']
data['Abdomen_to_Hip_Ratio'] = data['Abdomen'] / data['Hip']
data['Chest_to_Abdomen_Ratio'] = data['Chest'] / data['Abdomen']

# Doing the same for the scaled data / the training set

X_train['Chest_to_Hip_Ratio'] = X_train['Chest'] / X_train['Hip']
X_train['Abdomen_to_Hip_Ratio'] = X_train['Abdomen'] / X_train['Hip']
X_train['Chest_to_Abdomen_Ratio'] = X_train['Chest'] / X_train['Abdomen']

# Display the updated DataFrame
print(data.head())
```

The ratio data obtained can help us make several assessments about the population in our dataset and hence a generalized understanding of the same demographic, i.e, men in general.

# Correlation Mapping

Now we will map the relations between features to further help with selecting fewer features. Futhermore as part of one of the secondary objectives of our project, this will also help identify which parts of our body might have similar adherence to fat deposition/distribution. 

```python
plt.figure(figsize=(10, 7))
sns.heatmap(X_train.corr(), annot=True, cmap="plasma");
```

Now lets narrow down the highly correlated features with a function that implements a threshold:

```python
def correlation(df, threshold):
    col_cor = set()
    cor_mat = df.corr()

    for i in range(len(cor_mat)):
        for j in range(i):
            if abs(cor_mat.iloc[i, j]) > threshold:
                col_name = cor_mat.columns[i]
                col_cor.add(col_name)

    return col_cor

ans = correlation(X_train, threshold=0.85)

print("Highly Correlated Features:")
for feature in ans:
    print(feature)
```

#### Let's study the correlation between these now:

```python
correlated_features = list(ans)

# Subset the DataFrame based on the highly correlated features
subset_df = X_train[correlated_features]

# Calculate the correlation matrix for the subset
correlation_matrix = subset_df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a heatmap using seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

# Show the plot
plt.title('Correlation Matrix of Highly Correlated Features')
plt.show()
```

So we can see that Thigh and Hip are similar and Chest and Abdomen are similar, So lets select only one of each, Thigh and Abdomen in this case. Hence making our final list of features Weight, Abdomen, Thigh and Wrist. 

## Training and Testing Our Model:

We are using 3 different classifiers to train our model. Here are the results of training and testing our model.

### 1. SVM:

```python
from sklearn.model_selection import train_test_split, cross_val_predict, StratifiedKFold
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, f1_score, roc_curve, auc, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# X contains the selected features and y is the target variable
X = data.drop(columns=[target_variable])
y = data[target_variable]
selected_features = ['Weight', 'Abdomen', 'Thigh', 'Wrist']
X_selected = X[selected_features]

# Convert y to discrete classes since it is continuous

y = (y > y.mean()).astype(int)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Initialize the SVM model
svm_model = SVC(probability=True, random_state=42)

# Perform stratified cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
y_train_pred = cross_val_predict(svm_model, X_train, y_train, cv=cv, method='predict_proba')[:, 1]

# Fit the model on the training set
svm_model.fit(X_train, y_train)

# Predictions on the test set
y_test_pred = svm_model.predict_proba(X_test)[:, 1]

# Calculate ROC curve and AUC for training set
fpr_train, tpr_train, thresholds_train = roc_curve(y_train, y_train_pred)
roc_auc_train = auc(fpr_train, tpr_train)

# Calculate ROC curve and AUC for test set
fpr_test, tpr_test, thresholds_test = roc_curve(y_test, y_test_pred)
roc_auc_test = auc(fpr_test, tpr_test)

# Visualize ROC curves
plt.figure(figsize=(8, 6))
plt.plot(fpr_train, tpr_train, label=f'Train AUC = {roc_auc_train:.2f}')
plt.plot(fpr_test, tpr_test, label=f'Test AUC = {roc_auc_test:.2f}')
plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()

# Confusion Matrix for the test set
y_test_pred_binary = svm_model.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_test_pred_binary)

# Visualize Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap='Blues', cbar=False)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()

# Calculate and print accuracy and F1 scores
accuracy_train = accuracy_score(y_train, (y_train_pred > 0.5).astype(int))
f1_train = f1_score(y_train, (y_train_pred > 0.5).astype(int))

accuracy_test = accuracy_score(y_test, (y_test_pred > 0.5).astype(int))
f1_test = f1_score(y_test, (y_test_pred > 0.5).astype(int))

print(f'Training Accuracy: {accuracy_train:.2f}')
print(f'Training F1 Score: {f1_train:.2f}')
print(f'Test Accuracy: {accuracy_test:.2f}')
print(f'Test F1 Score: {f1_test:.2f}')
```

# FOCUS 2: BODY TYPE IDENTIFICATION:

Now we will be using our engineered features to understand the common body types in the male population as reflection of the results from the test population. A way to identify trends in body type will be to use the body measurement ratio features we engineered and perform cluster analysis on them to narrow down possible trends. We decided to use K-means clustering as it can group similar data points together based on certain features, helping us identify distinct types.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Importing the 3D plotting module

# Features of interest for clustering
cluster_features = ['Chest_to_Hip_Ratio', 'Abdomen_to_Hip_Ratio', 'Chest_to_Abdomen_Ratio']

# Extracting the relevant features for clustering
cluster_data = data[cluster_features]

# Scaling the data using MinMaxScaler
scaler = MinMaxScaler() 
cluster_data_scaled = scaler.fit_transform(cluster_data)

# Determine the optimal number of clusters using the Elbow Method
wcss = []  # Within-Cluster-Sum-of-Squares

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(cluster_data_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Choose the optimal number of clusters
optimal_k = 3

# Apply K-means clustering
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42)
cluster_labels = kmeans.fit_predict(cluster_data_scaled)

# Adding cluster labels to the dataset
data['Cluster'] = cluster_labels

# Visualize the clusters in 3D
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(cluster_data_scaled[:, 0], cluster_data_scaled[:, 1], cluster_data_scaled[:, 2], c=cluster_labels, cmap='viridis')

ax.set_xlabel('Chest-to-Hip Ratio (Normalized)')
ax.set_ylabel('Abdomen-to-Hip Ratio (Normalized)')
ax.set_zlabel('Chest-to-Abdomen Ratio (Normalized)')

# Add a legend
legend = ax.legend(*scatter.legend_elements(), title="Clusters")
ax.add_artist(legend)

plt.title('Clusters in 3D Space')
plt.show()

# Calculate mean or median values for each cluster
cluster_means = data.groupby('Cluster').mean()[cluster_features]
cluster_medians = data.groupby('Cluster').median()[cluster_features]

print("Cluster Means:")
print(cluster_means)

print("\nCluster Medians:")
print(cluster_medians)

```

Great! We have obtained the ratios for three distinct populations and now we can categorize them on the basis of body type:

1. Population 1: Larger Hips in comparison to chest and abdomen: This is an indicator of the gynoid or pear-shaped body type.

2. Population 2: Larger Chest and similar sized abdomen  in comparison to hips. This is a Trapezoid or torch body type.

3. Population 3: Larger hips in comparison to chest and abdomen but larger chest in comparison to abdomen. This is an hourglass body type.

These body types have been identified with the help of the following research paper: 

Kousar, S., Ramzan, M. B., &amp; et. al. (n.d.). Classification of male upper body shape: An innovative approach. Journal of Engineered Fibers and Fabrics. https://journals.sagepub.com/doi/full/10.1177/15589250231177447 

# FOCUS 3: AGING AND ITS IMPLICATIONS ON BODY FAT

Now we can focus on understanding the relationship between aging and overall body fat. It would be great to gain insight on whether aging can being predisposed to putting on more or less fat. We train a linear regression model to understand these trends. 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

# X_age is the feature (Age) and y is the target (BodyFat)
X_age = data[['Age']]
y_bodyfat = data['BodyFat']

# Checking for NaN values in y
nan_indices = y_bodyfat.isna()
X_age = X_age[~nan_indices]
y_bodyfat = y_bodyfat[~nan_indices]

# Applying log transformation to y
y_bodyfat = np.log1p(y_bodyfat)

# Standardize the features and target variable
scaler_age = StandardScaler()
X_age_scaled = scaler_age.fit_transform(X_age)

scaler_ratio = StandardScaler()
y_bodyfat_scaled = scaler_ratio.fit_transform(y_bodyfat.values.reshape(-1, 1))

# Initializing the Random Forest Regressor
model_rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Fitting the model to the data
model_rf.fit(X_age_scaled, y_bodyfat_scaled)

# Make predictions using the trained model
predicted_bodyfat_scaled_rf = model_rf.predict(X_age_scaled)

# Inverse transform the predicted bf to the original scale after log transformation
predicted_bodyfat_rf = np.expm1(scaler_ratio.inverse_transform(predicted_bodyfat_scaled_rf.reshape(-1, 1)))

# Plot the actual vs. predicted bf for Random Forest
plt.figure(figsize=(10, 6))
plt.scatter(X_age, np.expm1(y_bodyfat), label='Actual Body Fat', alpha=0.7)
plt.scatter(X_age, predicted_bodyfat_rf, label='Predicted Body Fat (Random Forest)', alpha=0.7)

# Linear Regression with Polynomial Features
model_poly = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
model_poly.fit(X_age_scaled, y_bodyfat_scaled)
predicted_bodyfat_scaled_poly = model_poly.predict(X_age_scaled)
predicted_bodyfat_poly = np.expm1(scaler_ratio.inverse_transform(predicted_bodyfat_scaled_poly.reshape(-1, 1)))
plt.scatter(X_age, predicted_bodyfat_poly, label='Predicted Body Fat (Poly Regression)', alpha=0.7)

# Support Vector Regression (SVR)
model_svr = SVR()
param_grid_svr = {'C': [0.1, 1, 10], 'epsilon': [0.01, 0.1, 0.2], 'kernel': ['linear', 'rbf']}
grid_search_svr = GridSearchCV(model_svr, param_grid_svr, cv=5)
grid_search_svr.fit(X_age_scaled, y_bodyfat_scaled)
best_svr = grid_search_svr.best_estimator_
predicted_bodyfat_scaled_svr = best_svr.predict(X_age_scaled)
predicted_bodyfat_svr = np.expm1(scaler_ratio.inverse_transform(predicted_bodyfat_scaled_svr.reshape(-1, 1)))
plt.scatter(X_age, predicted_bodyfat_svr, label='Predicted Body Fat (SVR)', alpha=0.7)

# Gradient Boosting Regressor
model_gb = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
model_gb.fit(X_age_scaled, y_bodyfat_scaled)
predicted_bodyfat_scaled_gb = model_gb.predict(X_age_scaled)
predicted_bodyfat_gb = np.expm1(scaler_ratio.inverse_transform(predicted_bodyfat_scaled_gb.reshape(-1, 1)))
plt.scatter(X_age, predicted_bodyfat_gb, label='Predicted Body Fat (Gradient Boosting)', alpha=0.7)

plt.title('Actual vs. Predicted Body Fat with Different Models')
plt.xlabel('Age')
plt.ylabel('Body Fat')
plt.legend()
plt.show()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_age_scaled, y_bodyfat_scaled, test_size=0.2, random_state=42)

# Random Forest on Training Set
model_rf.fit(X_train, y_train)
train_predictions_scaled_rf = model_rf.predict(X_train)
train_predictions_rf = np.expm1(scaler_ratio.inverse_transform(train_predictions_scaled_rf.reshape(-1, 1)))
train_mse_rf = mean_squared_error(y_train, train_predictions_rf)
train_mae_rf = mean_absolute_error(y_train, train_predictions_rf)
print(f"Training MSE (Random Forest): {train_mse_rf:.4f}")
print(f"Training MAE (Random Forest): {train_mae_rf:.4f}")

# Polynomial Regression on Training Set
model_poly.fit(X_train, y_train)
train_predictions_scaled_poly = model_poly.predict(X_train)
train_predictions_poly = np.expm1(scaler_ratio.inverse_transform(train_predictions_scaled_poly.reshape(-1, 1)))
train_mse_poly = mean_squared_error(y_train, train_predictions_poly)
train_mae_poly = mean_absolute_error(y_train, train_predictions_poly)
print(f"Training MSE (Poly Regression): {train_mse_poly:.4f}")
print(f"Training MAE (Poly Regression): {train_mae_poly:.4f}")

# SVR on Training Set
best_svr.fit(X_train, y_train)
train_predictions_scaled_svr = best_svr.predict(X_train)
train_predictions_svr = np.expm1(scaler_ratio.inverse_transform(train_predictions_scaled_svr.reshape(-1, 1)))
train_mse_svr = mean_squared_error(y_train, train_predictions_svr)
train_mae_svr = mean_absolute_error(y_train, train_predictions_svr)
print(f"Training MSE (SVR): {train_mse_svr:.4f}")
print(f"Training MAE (SVR): {train_mae_svr:.4f}")

# Gradient Boosting on Training Set
model_gb.fit(X_train, y_train)
train_predictions_scaled_gb = model_gb.predict(X_train)
train_predictions_gb = np.expm1(scaler_ratio.inverse_transform(train_predictions_scaled_gb.reshape(-1, 1)))
train_mse_gb = mean_squared_error(y_train, train_predictions_gb)
train_mae_gb = mean_absolute_error(y_train, train_predictions_gb)
print(f"Training MSE (Gradient Boosting): {train_mse_gb:.4f}")
print(f"Training MAE (Gradient Boosting): {train_mae_gb:.4f}")
```

As we can see the best model for this body fat prediction for different ages is Polynomial Regression. 

As for insights, from looking at the scatter plot we can tell that there does seem to be an increasing trend of increase in overall body fat with age. We acknowledge that this might be due to an array of reasons, but regardless, this is important to identify.
